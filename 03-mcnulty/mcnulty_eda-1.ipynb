```python
import os
import re
import pandas as pd
import numpy as np
import datetime as dt
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.decomposition import PCA
```

```python
pat = re.compile('loan_cleaned_')
fname = sorted([x for x in os.listdir('data') if pat.match(x)])[-1]
df = (pd.read_csv('data/%s' % fname)
      .set_index('id'))
```

```python
df.head()
```

```python
sns.scatterplot(x='loan_amnt', y='int_rate', hue='default', data=df)
plt.show()
```

```python
sns.scatterplot('int_rate', 'annual_inc', 'default', data=df)
plt.show()
```

```python
sns.scatterplot('open_acc', 'dti', 'default', data=df)
plt.show()
```

```python
bal_mask = df.revol_bal < 1000000
util_mask = df.revol_util < 800
sns.scatterplot('revol_bal', 'revol_util', 'default', data=df.loc[bal_mask & util_mask, ])
plt.show()
```
