```python
url = 'https://webrobots.io/kickstarter-datasets/'
response = requests.get(url)
page = response.text
```

```python
soup = BeautifulSoup(page,"html5lib")
```

```python
links = soup.find_all('a',text='CSV')
links
```

```python
for link in links:
    !wget {link['href']} -P /home/brendanfitzpatrick/projects/04-fletcher/files
    for name in glob.glob('/home/brendanfitzpatrick/projects/04-fletcher/files/*.zip'):
        !unzip {name} -d /home/brendanfitzpatrick/projects/04-fletcher/files/
    for name in glob.glob('/home/brendanfitzpatrick/projects/04-fletcher/files/*.csv'):
        temp = pd.read_csv(name)
        ks = pd.concat([ks, temp])
        ks.drop_duplicates(inplace=True)
    for name in glob.glob('/home/brendanfitzpatrick/projects/04-fletcher/files/*'):
        !rm {name}
```
